```python


#Program to Find area of rectangle
length=float(input("Enter length:"))
breadth=float(input("Enter breadth:"))

#Calculating the area
area=length*breadth

#Display the result
print("Area of rectangle=",area)

#Program to calculate simple interest
p=float(input("Enter principal amount:"))
r=float(input("Enter rate of interest:"))
t=float(input("Enter time(in years):"))

#Calculating the simple interest
si=(p*r*t)/100

#Display the result
print("Simple interest=",si)

#Program to convert temperature from celsius to fahrenheit
celsius=float(input("Enter temperature in celsius:"))

#Formula to convert temperature from celsius to fahrenheit
fahrenheit=(celsius*9/5)+32

#Display the result
print("Temperature in fahrenheit=",fahrenheit)

#Program to calculate the average of 3 numbers
num1=float(input("Enter first number:"))
num2=float(input("Enter second number:"))
num3=float(input("Enter third number:"))

#Formula to calculate average
average=(num1+num2+num3)/3

#Display the result
print("Average=",average)

#Find Square and Cube of a number
num=int(input("Enter a number:"))

square=num**2
cube=num**3

#Display the result
print("Square=",square)
print("Cube=",cube)

#Program to swap two numbers without third variable
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))

print("Before swapping")
print("a=",a)
print("b=",b)

a=a+b
b=a-b
a=a-b

print("After swapping")
print("a=",a)
print("b=",b)

#Program to student report
#Taking student details
name=input("Enter student name:")
roll_no=input("Enter roll number:")

#Taking the marks of 3 subjects
marks1=float(input("Enter marks of subject1:"))
marks2=float(input("Enter marks of subject2:"))
marks3=float(input("Enter marks of subject3:"))

#Calculating total
total=marks1+marks2+marks3

#Calculating percentage
percentage=total/3

#Display the report
print("\n-----Student Report-----")
print("Student Name:",name)
print("Roll Number:",roll_no)
print("Marks in Subject1:",marks1)
print("Marks in Subject2:",marks2)
print("Marks in Subject3:",marks3)
print("Total Marks:",total)
print("Percentage:",percentage)
```
